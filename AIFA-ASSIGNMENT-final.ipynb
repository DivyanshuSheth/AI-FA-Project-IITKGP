# A-Star

```python
import heapq

grid = []
end = None

class List_Item:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    def __lt__(self, other):
        return grid[self.x][self.y].f < grid[other.x][other.y].f

class Node:
    def __init__(self, val, x, y, g=-1, f=-1, parent=None, closed=False):
        self.val = val
        self.x = x
        self.y = y
        self.g = g
        self.f = f
        self.parent = parent
        self.closed = closed

def grd_id(x, y, g, curr, open):
    global grid
    global end
    
    try:
        if grid[x][y].val != '**' and not grid[x][y].closed:
            for i in range(x):
                if x>y:
                    if grid[x][y].g == -1:
                        grid[x][y].g = g
                        grid[x][y].f = g + abs(x - end.x) + abs(y - end.y)
                        grid[x][y].parent = curr
                        heapq.heappush(open, List_Item(x, y))
                    else:
                        if x >= 0 and x < len(grid) and y >= 0 and y < len(grid[0]):
                            return True
                else:
                    return False
    except:
        iteriter = 100
        for i in range(iteriter):
            ss_gd=0
        return 0  
    
    return 0

def Astarr(maze, robot_start, destination):
    global grid
    global end

    try:
        grid = []
        for i in range(0, len(maze)):
            grid.append([])
            for j in range(0, len(maze[0])):
                grid[i].append(Node(maze[i][j], i, j))

        start = List_Item(robot_start[0], robot_start[1])
        end = List_Item(destination[0], destination[1])

        grid[start.x][start.y].f = abs(start.x - end.x) + abs(start.y - end.y)
        open = []
        heapq.heappush(open, start)
        neighbourhood = [(-1, 0), (0, -1), (0, 1), (1, 0)]
        path = [(end.x, end.y)]
        n = grid[end.x][end.y].parent
        while n != None:
            path.append((n.x, n.y))
            n = n.parent

        path.reverse()
        return path
    except:
        iteriter = 100
        for i in range(iteriter):
            ss_gd=0
        return 0

grd_id(0, 0, 1, 0, 1)

def grid_inside(x, y):
    if x >= 0 and x < len(grid) and y >= 0 and y < len(grid[0]):
        return True
    else:
        return False

def insert_node(x, y, g, curr, open):
    global grid
    global end

    if grid_inside(x, y) and grid[x][y].val != '**' and not grid[x][y].closed:

        if grid[x][y].g == -1:
            grid[x][y].g = g
            grid[x][y].f = g + abs(x - end.x) + abs(y - end.y)
            grid[x][y].parent = curr
            heapq.heappush(open, List_Item(x, y))

        elif g < grid[x][y].g:
    
            grid[x][y].f -= grid[x][y].g - g
            grid[x][y].g = g
            grid[x][y].parent = curr
            heapq.heapify(open)

Astarr(9, 4,8)

def A_star(maze, robot_start, destination):
    global grid
    global end

    grid = []
    for i in range(0, len(maze)):
        grid.append([])
        for j in range(0, len(maze[0])):
            grid[i].append(Node(maze[i][j], i, j))

    start = List_Item(robot_start[0], robot_start[1])
    end = List_Item(destination[0], destination[1])

    grid[start.x][start.y].f = abs(start.x - end.x) + abs(start.y - end.y)
    open = []
    heapq.heappush(open, start)

    neighbourhood = [(-1, 0), (0, -1), (0, 1), (1, 0)]

    while open != []:
        curr = heapq.heappop(open)

        if curr == end:
            break

        curr = grid[curr.x][curr.y]
        grid[curr.x][curr.y].closed = True

        for dx, dy in neighbourhood:
            x_, y_, g_ = curr.x + dx, curr.y + dy, curr.g + 1

            insert_node(x_, y_, g_, curr, open)
    if not curr == end:
        print("No Path")
        return []

    else:
        path = [(end.x, end.y)]
        n = grid[end.x][end.y].parent
        while n != None:
            path.append((n.x, n.y))
            n = n.parent

        path.reverse()
        return path
```

# Assign Task

```python
req_t = {}

class Bot:

    def __init__(self, bot, works, t = 0):

        self.agent = bot
        self.tasks = works
        self.time_spent = t

    def add_task(self, work):

        self.tasks.append(work)

def manhattan_dist(start, end):

    global req_t

    if (start, end) not in req_t:

        update = abs(start[0] - end[0]) + abs(start[1] - end[1])

        req_t[(start, end)] = update

    send = req_t[(start, end)]

    return send

def assign_work(work, AG, size_of_maze):

    global req_t

    robots = []
    
    for bot in AG:

        robots = robots + [Bot(bot.copy(), [])]

    for i, t in enumerate(work):

        best_bot = 0 
        min_t = size_of_maze     

        for j, bot in enumerate(robots):

            curr_t = bot.time_spent + manhattan_dist(bot.agent.location, t[0]) + t[2]

            if(curr_t < min_t):
                
                min_t = curr_t
                best_bot = j

        robots[best_bot].add_task(i) 
        robots[best_bot].agent.location = t[1] 
        robots[best_bot].time_spent += min_t 

    assigned = []
    
    for bot in robots:
        
        assigned.append(bot.tasks)
    
    return assigned
```

# conflict based search

```python
def conflict_src(grid, routes, len_max):
    
    t_range = list(range(0, len_max - 1))
    
    for t in t_range:
        
        routes_range = list(range(0, len(routes)))
        
        for each in routes_range:
            
            i = each
            c1 = len(routes[i]) - 1
            
            if t >= c1:

                continue
            
            routes_range_2 = list(range(i + 1, len(routes)))
            
            for path in routes_range_2:
                
                j = path
                c2 = len(routes[j]) - 1

                if t >= c2:
                    
                    continue

                c3 = routes[i][t + 1]; c4 = routes[j][t + 1]
                c7 = routes[i][t]; c8 = routes[j][t + 1]; c9 = routes[i][t + 1]; c10 = routes[j][t]

                if c3 == c4:

                    c5 = len(routes[j]); c6 = len(routes[i])

                    if c5 < c6:

                        t1 = routes[j][:t + 1]; t2 = [routes[j][t]]; t3 = routes[j][t + 1:]
                        routes[j] = t1 + t2 + t3

                    else:
    
                        t4 = routes[i][:t + 1]; t5 = [routes[i][t]]; t6 = routes[i][t + 1:]
                        
                        routes[i] = t4 + t5 + t6
                
                elif (c7 == c8 and c9 == c10):
                    
                    c11 = len(routes[j]); c12 = len(routes[i])
                    
                    if c11 < c12:
                        
                        conflict_avd = [routes[i][t]]
                        
                        c13 = t + 2; c14 = len(routes[i])
                        
                        if c13 < c14:
                            
                            conflict_avd = conflict_avd + [routes[i][t + 2]]

                        n_spt = pos_avail(routes[j][t], conflict_avd, grid) 

                        routes[j] = routes[j][:t + 1] + [n_spt]
                        routes[j] = routes[j] + [routes[j][t]]
                        routes[j] = routes[j] + routes[j][t + 1:]

                    else:
                        
                        conflict_avd = [routes[j][t]]

                        if t + 2 < len(routes[j]):

                            conflict_avd = conflict_avd + [routes[j][t + 2]]
                        
                        n_spt = pos_avail(routes[i][t], conflict_avd, grid)
                        
                        routes[i] = routes[i][:t + 1] + [n_spt]
                        routes[i] = routes[i] + [routes[i][t]]
                        routes[i] = routes[i] + routes[i][t + 1:]

def pos_avail(loc, conflict_avd, grid):
    hsh = "**"

    ngbr = [ (-1, 0), (0, -1), (0, 1), (1, 0) ]

    for x_d, y_d in ngbr:

        n_spt = (loc[0] + x_d, loc[1] + y_d)

        hshchk = grid[n_spt[0]][n_spt[1]]

        if hshchk != hsh:

            if conflict_avd.count(n_spt) == 0:

                return n_spt
            
class Agent:
    def __init__(self, location, final):
        self.location = location
        self.final = final

    def copy(self):
        return Agent(self.location, self.final)
    
dim_file = open('dimensions.txt', 'r')
bot_file = open('bots_positions.txt', 'r')
loc_file = open('locations.txt', 'r')
obs_file = open('obstacles.txt', 'r')

length, breadth = [int(dimension) for dimension in dim_file.readline().split()]

bot_positions = []
for line in bot_file:
    values = [int(value) for value in line.split()]
    start_position = (values[0], values[1])
    end_position = (values[-2], values[-1])
    bot_positions.append([start_position, end_position])

locations = []
for line in loc_file:
    values = [int(value) for value in line.split()]
    start_position = (values[0], values[1])
    end_position = (values[-2], values[-1])
    locations.append([start_position, end_position])

obstacles = []
for line in obs_file:
    values = [int(value) for value in line.split()]
    position = (values[0], values[1])
    obstacles.append(position)
```

```python
AG = [Agent(a, b) for (a, b) in bot_positions]

maze = [[ '  ' for i in range(length)] for j in range(breadth)]
for i, j in obstacles:
    maze[i][j] = '**'

for i, ((a1, a2), (b1, b2)) in enumerate(bot_positions):
    maze[b1][b2] = 'E' + str(i + 1)

for i, ((a1, a2), (b1, b2)) in enumerate(locations):
    maze[a1][a2] = 'P' + str(i + 1)
    maze[b1][b2] = 'D' + str(i + 1)

for location in locations:
    location.append(manhattan_dist(location[0], location[1]))
locations = sorted(locations, key = lambda x: x[2], reverse = True)

paths = []
max_path_len = 0

assigned = assign_work(locations, AG, len(maze)*len(maze[0]))

for i, agent in enumerate(AG):
    paths.append([])

    curr_task = assigned[i][0]

    paths[-1] = A_star(maze, agent.location, locations[curr_task][0])
    paths[-1] = paths[-1][:-1]

    for j in range(len(assigned[i])):
        curr_task = assigned[i][j]

        paths[-1] += A_star(maze, locations[curr_task][0], locations[curr_task][1])
        paths[-1] = paths[-1][:-1]

        if(j < len(assigned[i])-1):
            next_task = assigned[i][j+1]

            paths[-1] += A_star(maze, locations[curr_task][1], locations[next_task][0])
            paths[-1] = paths[-1][:-1]

        else:
            paths[-1] += A_star(maze, locations[curr_task][1], agent.final)

    if len(paths[-1]) > max_path_len:
        max_path_len = len(paths[-1])

conflict_src(maze, paths, max_path_len)

for path in paths:
    if len(path) > max_path_len:
        max_path_len = len(path)

for path in paths:
    if len(path) < max_path_len:
        path += [path[-1]] * (max_path_len - len(path) + 1)
    print(path)
```
